```python
def isPrime(no):
    if no==1:
        return False
    sqrt=no ** 0.5
    n=0
    for k in range(2,int(sqrt)+1):
        if no%k==0:
            n=n+1
            break
    if n==0:
        return True
    return False
def findLen(num):
    
    lenght=0
    while num > 0:
        
        lenght+=1
        num=int(num/10)
    return lenght

def leftshift(num):
    if(num<10):
        return num
    else:
        l=[]
        num=str(num)
        for i in num:
            l.append(i)
        
        del l[0]
        
        num=int(''.join(l))
        return num
    
def Rightshift(num):
    if(num<10):
        return num
    else:
        l=[]
        num=str(num)
        for i in num:
            l.append(i)
        
        del l[-1]
        
        num=int(''.join(l))
        return num


    

lst=[]
Lshift=0
Rshift=0
count=0
flst=[]
sum=0
for i in range(11,1000000):
    if isPrime(i)==True:
        lst.append(i)
        
for i in range(0,len(lst)):
        
    Lshift=lst[i]
    Rshift=lst[i]
    lenght=findLen(lst[i])
    for j in range(lenght):
        Lshift=leftshift(Lshift)
        Rshift=Rightshift(Rshift)
        if isPrime(Lshift)==True and isPrime(Rshift)==True:
            count+=1
        else:
            break
            
    if(lenght==count):
        flst.append(lst[i])
    count=0
    if(len(flst)==13):
        break
        
print(flst)
for i in range(len(flst)):
    sum=sum+flst[i]
print("The sum is: ",sum)
            
```
