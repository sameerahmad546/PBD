```python
def isPrime(no):
    sqrt=no ** 0.5
    n=0
    for k in range(2,int(sqrt)+1):
        if no%k==0:
            n=n+1
            break
    if n==0:
        return True
    return False

def givesum(lst,pstart,pend):
    sum=0
    for i in range(pstart,pend):
        sum=sum+lst[i]
        
    return sum



lst=[]
start=-1
end=1
start1=0
end1=1
sum=0
Fstart=0
Fend=1
lenght=0
lenght2=0
for i in range(2,1000):
    if isPrime(i)==True:
        lst.append(i)
        
for l in range(0,100):
    start+=1
    end=start+1
    sum=0
    lenght=0
    while sum<1000:
        sum=0
        sum=givesum(lst,start,end)
        
        if isPrime(sum)==True and sum<1000:
            start1=start
            end1=end
            lenght=end1-start1
        end+=1


    
    if lenght > lenght2:
        lenght2=lenght
        Fstart=start1
        Fend=end1


sum=givesum(lst,Fstart,Fend)
for l in range(Fstart,Fend):
    print(lst[l])

print("Final sum",sum)
print(lenght2)






```
